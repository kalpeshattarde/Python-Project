### 1️⃣ Importing the Required Libraries

```python
import bs4
from bs4 import BeautifulSoup      #to parse the HTML and extract data
import requests                    #to send a request to a website and download HTML
import pandas as pd                #to store the extracted data into a dataframe
```

### 2️⃣ Finding the URL Containing Nifty 50 Stock Data

```python
url='https://www.screener.in/screens/281185/nifty-50'
page=requests.get(url)
page
```

### 3️⃣ Sending Request to the Website

```python
soup=BeautifulSoup(page.text)
soup
```

### 4️⃣ Extracting the Required Stock Data

```python
table=soup.find("table")
table=soup.find("tbody")
table
```

```python
rows = table.find_all("tr")
rows
```

### 5️⃣ Storing data into the list 

```python
Nifty_50=[]

for i in rows:
    cols=i.find_all("td")
    #print(cols)
    stocks = [P.text for P in cols]
    #print(car_sales)
    Nifty_50.append(stocks)

print(Nifty_50)
```

```python
Nifty_50
```

```python
Nifty_50.pop(0)
Nifty_50.pop(15)
```

```python
Nifty_50
```

```python
Sr_no = []
Name = []
Share_price = []
PE_Ratio = []
Market_Capital = []
Dividend_Yeild = []
Last_QTR_Profit = []
YoY_Profit_Growth = []
Last_QTR_Sales = []
QTR_Sales_Growth = []
Return_on_Capital = []

for i in Nifty_50:
    Sr_no.append(i[0])
    Name.append(i[1].strip())
    Share_price.append(i[2])
    PE_Ratio.append(i[3])
    Market_Capital.append(i[4])
    Dividend_Yeild.append(i[5])
    Last_QTR_Profit.append(i[6])
    YoY_Profit_Growth.append(i[7])
    Last_QTR_Sales.append(i[8])
    QTR_Sales_Growth.append(i[9])
    Return_on_Capital.append(i[10])
    
```

```python
Sr_no
```

```python
Data = {
    "Sr No": Sr_no,
    "Company Name": Name,
    "Share Price (Rs.)": Share_price,
    "P/E Ratio": PE_Ratio,
    "Market Capital (Rs. Cr.)": Market_Capital,
    "Dividend Yield (%)": Dividend_Yeild,
    "Last QTR Profit (Rs. Cr.)": Last_QTR_Profit,
    "YoY Profit Growth (%)": YoY_Profit_Growth,
    "Last QTR Sales (Rs. Cr.)": Last_QTR_Sales,
    "QTR Sales Growth (%)": QTR_Sales_Growth,
    "Return on Capital (%)": Return_on_Capital
}
```

```python
print(Data)
```

### 6️⃣ Converting Extracted Data Into a DataFrame

```python
Nifty_50_info = pd.DataFrame(Data)

Nifty_50_info
```

### 7️⃣ Saving the Data

```python
Nifty_50_info.to_csv("Nifty50 stocks list.csv")
```
