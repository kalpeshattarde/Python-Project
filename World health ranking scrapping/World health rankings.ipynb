### 1️⃣ Importing the Required Libraries

```python
import bs4
from bs4 import BeautifulSoup      #to parse the HTML and extract data
import requests                    #to send a request to a website and download HTML
import pandas as pd                #to store the extracted data into a dataframe
```

### 2️⃣ Finding the URL Containing World Population Review Data

```python
url='https://worldpopulationreview.com/country-rankings/healthiest-countries'
page=requests.get(url)
page
```

### 3️⃣ Sending Request to the Website

```python
soup=BeautifulSoup(page.text)
soup
```

### 4️⃣ Extracting the Required Rank Data

```python
table=soup.find("table")
table=soup.find("tbody")
table
```

```python
rows = table.find_all("tr")
rows
```

### 5️⃣ Storing data into the list 

```python
data =[]

for i in rows:
    cols=i.find_all("td")
    #print(cols)
    rank = [P.text for P in cols]
    #print(car_sales)
    data.append(rank)

print(data)
```

```python
World_Health = []

for row in data:
    row = row[1:]
    row = [None if x == '' else x for x in row]
    World_Health.append(row)

for r in World_Health[:5]:
    print(r)
```

```python
World_Health
```

```python
COUNTRY = []
GLOBAL_HEALTH_INDEX_2024 = []
LEGATUM_PROSPERITY_INDEX_HEALTH_RANKINGS_2023 = []
GLOBAL_HEALTH_SECURITY_INDEX_2021 = []
MONEY_UK_HEALTHIEST_RANK_2021 = []
BLOOMBERG_GLOBAL_HEALTH_INDEX_2021 = []

for i in World_Health:
    COUNTRY.append(i[0])
    GLOBAL_HEALTH_INDEX_2024.append(i[1])
    LEGATUM_PROSPERITY_INDEX_HEALTH_RANKINGS_2023.append(i[2])
    GLOBAL_HEALTH_SECURITY_INDEX_2021.append(i[3])
    MONEY_UK_HEALTHIEST_RANK_2021.append(i[4])
    BLOOMBERG_GLOBAL_HEALTH_INDEX_2021.append(i[5])
```

```python
COUNTRY
```

```python
Data = pd.DataFrame({
    'COUNTRY': COUNTRY,
    'GLOBAL HEALTH INDEX 2024': GLOBAL_HEALTH_INDEX_2024,
    'LEGATUM PROSPERITY INDEX HEALTH RANKINGS 2023': LEGATUM_PROSPERITY_INDEX_HEALTH_RANKINGS_2023,
    'GLOBAL HEALTH SECURITY INDEX 2021': GLOBAL_HEALTH_SECURITY_INDEX_2021,
    'MONEY UK HEALTHIEST RANK 2021': MONEY_UK_HEALTHIEST_RANK_2021,
    'BLOOMBERG GLOBAL HEALTH INDEX 2021': BLOOMBERG_GLOBAL_HEALTH_INDEX_2021
})
```

```python
print(Data)
```

### 6️⃣ Converting Extracted Data Into a DataFrame

```python
World_Health_Surveys = pd.DataFrame(Data)

World_Health_Surveys
```

### 7️⃣ Saving the Data

```python
World_Health_Surveys.to_csv("World Health Surveys.csv")
```
